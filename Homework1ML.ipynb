**33\.** Как получить последние n строк датафрейма с суммой строк 100?  (★★★)

```python
df = pd.DataFrame(np.random.randint(1, 30, 60).reshape(20,-1), columns=list('abc'))
sum = df.sum(1)
ans = []
totalsum = 0
for ind in np.flip(sum.index):
    if totalsum > 100:
        break
    else:
        ans.append(ind)
        totalsum+=sum[ind]
df.iloc[ans[::-1]]
```

```python
s = pd.Series(np.logspace(-2, 2, 30))
def iqr(series):
    q1, q3 = np.quantile(series, [0.25, 0.75])
    return q3 - q1

s[(s > q1 - 1.5*iqr(s)) & (s < q3 + 1.5*iqr(s))]
```

```python
df = df.drop(['Model', 'Make'], 1)
cat = cat.drop(['Model', 'Make'])
df
```

```python
more10 = df[cat].nunique().ge(10)
for c in cat[more10]:
    ca = df[c].copy()
    ca[~ca.isin(ca.value_counts().head(10).index)] = 'other'
    df[c] = ca
df
```

```python
numb = df.dtypes[df.dtypes != 'object'].index
missing = df.isna().apply(lambda x: x.sum()/len(df))
missing
```

```python
for x in df[numb]:
    d, u = np.nanquantile(df[x], [0.05, 0.95])
    df.loc[(df[x] < d), x] = d
    df.loc[(df[x] > u), x] = u
df
```

```python
df[numb] = df[numb].fillna(df[numb].mean())
df[cat] = df[cat].fillna('other')
df
```

# Regular expressions with Pandas

**1\.** Выделить в отдельную колонку event_id первую цифру в кавычках, во вторую колонку event_probability вероятность (★☆☆) 

```python
import re
df=pd.DataFrame({'col1':["{'34': 0.9839372122311848}",
                         "{'77': 0.6724645988404411}",
                         "{'77': 0.6491011266207006}",
                         "{'77': 0.6123085784265466}",
                         "{'34': 0.87}",
                         "{'77': 0.6225699652019304}",
                         "{'3': 0.5470808226524633}",
                         "{'77': 0.6233849004135044}",
                         "{'77': 0.7050888998743866}",
                         "{'52': 0.8945052223205372}"]})

df['event_id'] = df['col1'].apply(lambda x: re.search('\d', x).group(0))
df['event_probability'] = df['col1'].apply(lambda x: re.search('\d+.\d+', x).group(0))
df
```
